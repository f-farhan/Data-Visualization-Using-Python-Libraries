## 3. Seaborn vs Matplotlib

In the previous module we have seen some basic plotting techniques using **matplotlib**.Now,if we’re looking for a simpler way to plot attractive charts,then we can use **Seaborn**. The following example will make it more clear:

**Generating the data:**

```python
r = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(r.randn(500, 6), 0)
```

**Plotting the data using Matplotlib:**

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

**Plotting the data using Seaborn:**

```python
import seaborn as sns
sns.set()

plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

So,we see plotting with **seaborn** looks much better.

## 4.Basic Seaborn functions

Plots are basically used for visualizing the relationship between variables. Those variables can be either be completely numerical or a category like a group, class or division. Let's see some plots of different categories:

### 4.1. Line Plot

We can plot a lineplot using the function **sns.lineplot()**

```python
import seaborn as sns; sns.set()

import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

ax = sns.lineplot(x="timepoint", y="signal", data=fmri)
```

### 4.2 Dist Plot & Lmplot

**Dist Plot:**

Seaborn **dist plot**  is used to plot a **histogram**

```python
# Importing libraries 
import numpy as np 
import seaborn as sns 


# Selecting style as white, 
# dark, whitegrid, darkgrid 
# or ticks 
sns.set(style="dark") 

# Generate a random univariate 
# dataset 
rs = np.random.RandomState(10) 
d = rs.normal(size=100) 

# Plot a simple histogram and kde 
# with binsize determined automatically 
sns.distplot(d, kde=True, color="m")
```

**Lmplot**

```python
import seaborn as sns 

sns.set(style="ticks") 

# Loading the dataset 
df = sns.load_dataset("anscombe") 

# Show the results of a linear regression 
sns.lmplot(x="x", y="y", data=df)
```

### 4.3 Barplot

At first let us load a dataset:

```python
# import the seaborn libaray 
import seaborn as sns 

# import done to avoid warnings 
from warnings import filterwarnings 

# reading the dataset 
df = sns.load_dataset('tips') 

# first five entries if the dataset 
df.head() 
```

```python
# set the backgroud stle of the plot 
sns.set_style('darkgrid') 

# plot the graph using the default estimator mean 
sns.barplot(x ='sex', y ='total_bill', data = df, palette ='plasma') 

# or 
import numpy as np 

# change the estimator from mean to standard devaition 
sns.barplot(x ='sex', y ='total_bill', data = df, 
			palette ='plasma', estimator = np.std) 
#palette is used to set the color of the plot 
#estimator is used as a statistical function for estimation within each categorical bin.
```

```python
sb.barplot(x = 'sex', y = 'total_bill', data = df, ci=False)
```

### 4.7 Pair Plot

We use **pair plots** when we generalize joint plots to datasets of larger dimensions.**Pair Plot** is very useful for exploring correlations between multidimensional data, when you'd like to plot all pairs of values against each other.

```python
df = sns.load_dataset('tips')
df.head()

```

```python
sns.pairplot(df, hue ="sex", palette ='coolwarm') 
```

Let's see another **example:**

```python
df = sns.load_dataset("iris")
df.head()
```

```python
sns.pairplot(df, hue='species', height=2.5);
```

### 4.8. Faceted Histogram

```python
df = sns.load_dataset('tips')
df.head()

```

Using Seaborn's **FacetGrid** we'll take a look at some data that shows the amount that restaurant staff receive in tips based on various indicator data:

```python
df['tip_pct'] = 100 * df['tip'] / df['total_bill']

grid = sns.FacetGrid(df, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

### 4.9. Jointplot

We can use **sns.jointplot** to show the joint distribution between different datasets, along with the associated marginal distributions:

```python
df = sns.load_dataset('tips')
df.head()

```

```python
sns.jointplot(x ='total_bill', y ='tip', data = df) 
```

Let's see the same example with different style:

```python
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", data=df, kind='hex')
```

**KDE Plot**

**KDE** shows the density where the points match up the most 


```python
sns.jointplot(x ='total_bill', y ='tip', data = df, kind ='kde') 
```
