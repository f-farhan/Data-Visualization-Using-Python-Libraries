## 2.Creating Dataframe

```python
df = pd.DataFrame({
    'Overall Rank': [94, 93, 91, 88, 87, 85, 82],
    'Age': [
        33,
        35,
        29,
        26,
        28,
        30,
        31
    ],
    'Matches Won in 2019': [
        21,
        30,
        18,
        16,
        13,
        7,
        22
    ],
    'Matches Lost in 2019': [
        9,
        0,
        12,
        14,
        17,
        23,
        8
    ],
    'Position': [
        'CF',
        'LW',
        'LW',
        'CAM',
        'CDM',
        'CB',
        'LM'
    ]
}, columns=['Overall Rank', 'Age', 'Matches Won in 2019', 'Matches Lost in 2019', 'Position'])
```

```python
df
```

`DataFrame` also have indexes. As we can see in the "table" above, pandas has assigned a numeric, autoincremental index automatically to each "row" in our DataFrame. In our case, we know that each row represents a country, so we'll just reassign the index:

```python
df.index = [
    'Lionel Messi',
    'Cristiano Ronaldo',
    'Neymar Jr',
    'Eden Hazard',
    'Luka Modric',
    'Dani Alves',
    'Pique',
]
```

## 3.Indexing,Selection & Slicing

```python
df
```

```python
df.loc['Neymar Jr'] # Returns all information for the selected element 
```

```python
df.iloc[-1] # Returns all information for the last element 
```

```python
df['Age'] #Returns all values of the selected column with index
```

#### if we want to see a more DataFrame-like format we can use the to_frame method:

```python
df['Age'].to_frame()
```

```python
df[['Overall Rank', 'Position']] #Returns all values of the selected column with index
```

```python
df[1:3] #Returns first two row values
```

```python
df.loc['Dani Alves'].to_frame() #Returns all information of the selected element
```

```python
df.loc['Cristiano Ronaldo': 'Luka Modric', 'Position'].to_frame() #Returns column values for the selected index range
```

```python
df.loc['Cristiano Ronaldo': 'Luka Modric', ['Age', 'Position']]
```

```python
df
```

## 5.Dropping Data

```python
df
```

#### We can drop specifc rows or columns using df.drop() function

```python
df.drop('Pique')
```

```python
df.drop(['Luka Modric','Neymar Jr'])
```

```python
df.drop(columns=['Overall Rank', 'Age'])
```

```python
df.drop(['Dani Alves', 'Pique'], axis=0) ## axis=0 for rows
```

```python
df.drop(['Matches Won in 2019', 'Matches Lost in 2019'], axis=1) ## axis=1 for columns
```

#### We can also write (axis = columns,axis=row) instead of (axis=0,axis=1)

```python
df.drop(['Dani Alves', 'Pique'], axis='rows') 
```

```python
df.drop(['Matches Won in 2019', 'Matches Lost in 2019'], axis='columns') 
```

## 7.Modifying Dataframe

### Adding a new Column

```python
ballondor = pd.Series(
    [6, 5, 1],
    index=['Lionel Messi', 'Cristiano Ronaldo', 'Luka Modric'],
    name='BallonD-or'
)
```

```python
df['BallonD-or'] = ballondor
```

```python
df
```

---
### Replacing values per column:

```python
df['Overall Rank'] = '?'
```

```python
df
```

### Renaming Columns:


```python
df.rename(
    columns={
        'Matches Won in 2019': 'MW-19',
        'Matches Lost in 2019': 'ML-19'
    }, index={
        'Lionel Messi': 'Messi',
        'Cristiano Ronaldo': 'Ronaldo',
        'Neymar Jr': 'Neymar'
    })
```

```python
df.rename(index=str.upper) #Converts the index names to uppercase letters
```

```python
df.rename(index=lambda x: x.lower()) #Converts the index names to lowercase letters
```

```python
df.append(pd.Series({
    'Age': 29,
    'Position': 'LB',
}, name='Marcelo')) ##Adds a new row with values
```

```python
df
```

```python
df.loc['Marcelo'] = pd.Series({'Age': 29, 'Position': 'CB'})
```

```python
df
```

```python
df.reset_index()
```
